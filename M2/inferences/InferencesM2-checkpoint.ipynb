```python
#Fonction servant à nettoyer la phrase entrée par l'utilisateur 
#Sortie la phrase coupée en une liste de mots 
def cleaning_sentence():
    sentence_input = ''

    #utilisateur entre une phrase
    while sentence_input == '' : 
        print('Veuillez entrer une phrase à étudier :\n')
        print('Exemple de phrase qui fonctionne : La vache et le lait. \nExemple de phrase ne fonctionnant pas (ici le mot est en majuscule sans raison) : La VACHE boit du lait.\n')
        sentence_input = input()

    #on enlève ces caractères  
    items_replace = [",", ";", ".", "!", "?"]
    
    clean_sentence = sentence_input
    for items in items_replace :
        clean_sentence = clean_sentence.replace(items, " ") 

    #on découpe la phrase dans une liste de string 
    sentence_split = clean_sentence.split()    
    
    return sentence_split
```

```python
#Class Node, Edge et Graph

class Node:
    def __init__(self, mot, pred, succ, identificateur, final_bool):
        self.mot = mot 
        self.pred = pred #id du noeud precedent 
        self.succ = succ #id du noeud successeur 
        self.identificateur = identificateur #unique dans notre graphe (mais différent de l'id de neo4j)
        self.final_bool = final_bool #mot fin de phrase alors True 
        
class Edge:
    def __init__(self,identificateur, relation, weight, node_start, node_end):
        self.identificateur = identificateur
        self.rel = relation
        self.weight = weight 
        self.n_start = node_start
        self.n_end = node_end 

class Graph:
    def __init__(self, n, e):
        self.nodes = n
        self.edges = e
```

```python
#creation du graphe depuis une phrase
#Entrée liste de mots (le retour de la fonction cleaning_sentence)
#Sortie le graphe et les listes des identificateurs de noeuds et arêtes
def create_graph(sentence):
    
     
    length = len(sentence)
    
    #creation du premier noeud et de l'arête entre le premier noeud et premier mot 
    nodes = [Node("_START", -1, 0, -1, False)]
    edges = [Edge(-1, "r_succ", None, -1, 0)]

    for i in range(length) : 

        #enregistrement des mots ou voir s'ils sont déjà dans les fichiers 
        if (presenceFichier(sentence[i]+'.txt')):
            print("\nFichier trouvé.")
        else:
            if (enregistrementFichier(sentence[i]) == False):
                print("\nArrêt de la recherche car le mot \""+sentence[i]+"\"n'existe pas dans la base Jeux de mots.\n")
                return None, None 
            
        #creation et inserer les noeuds et arêtes dans le graphe 
        if i != length-1 : 
            nodes.append(Node(sentence[i], i-1, i+1, i, False))
            edges.append(Edge(i, "r_succ",None, i, i+1))
        else : 
            #dernier mot relié au noeud _END
            nodes.append(Node(sentence[i], i-1, -2, i, True))
            edges.append(Edge(i, "r_succ",None, i, -2))
        
    
    #creation du dernier noeud
    end_node = Node("_END", length-1, -2, -2, False)
    nodes.append(end_node)

   
    graph = Graph(nodes, edges)
    
    #liste des id des nodes et arêtes (pour l'instant pas d'intérêt)
    nodes_id = list(range(length))
    edges_id = list((range(-1, length-1)))+[-2]
    
    return graph, [nodes_id, edges_id]
```

```python
#creation graphe dans Neo4j 
#Entrée : le graphe créé à partir de la fonction create_graph
def create_neo4jGraph(graph): 
    
    #nettoyer le graphe neo4j 
    query_delete = "MATCH (n) DETACH DELETE (n)"
    result = graph_neo4j.run(query_delete)
    
    nodes_list = graph.nodes 
    edges_list = graph.edges
    
    #Creation des noeuds dans neo4j
    for node in nodes_list : 

        query_node = "CREATE (n:Noeud { mot : \""+node.mot+"\", pred :"+str(node.pred)+", succ : "+str(node.succ)+", id_mot : "+str(node.identificateur)+", final_bool : "+str(node.final_bool)+"})"
        result = graph_neo4j.run(query_node)
    
    #Creation des arêtes dans neo4j
    for edge in edges_list : 
        
        #Si on veut voir les arêtes en détails alors on les transforme en noeud dans neo4j pour avoir leur poids
        #on impose weight à 10
        #query_edge = "CREATE (e:Edge {rel:\"r_succ\", weight:10, n_start:"+n_start+", n_end:"+n_end+"})"
        
        #les arêtes en arête dans neo4j mais on perd l'information du poids 
        query_edge = "MATCH (n:Noeud {id_mot:"+str(edge.n_start)+"}), (n1:Noeud {id_mot:"+str(edge.n_end)+"}) CREATE (n)-[:R_SUCC]->(n1)"
        result = graph_neo4j.run(query_edge)

    
    
```
